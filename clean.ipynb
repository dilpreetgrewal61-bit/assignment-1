```python
# clean_shopping.ipynb
print("=== Shopping List Manager ===")

# Using 5 fixed slots (since lists not allowed)
items = ["", "", "", "", ""]
count = 0
running = True

while running:
    print("\n1. Add item")
    print("2. View items")
    print("3. Remove item")
    print("4. Count items")
    print("5. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        added = False
        for i in range(5):
            if items[i] == "":
                items[i] = input("Enter item: ")
                count += 1
                print("Item added!")
                added = True
                break
        if not added:
            print("Shopping list is full (max 5 items).")

    elif choice == "2":
        print("Current items:")
        for i in range(5):
            if items[i] != "":
                print(items[i])

    elif choice == "3":
        remove = input("Enter item to remove: ")
        removed = False
        for i in range(5):
            if items[i] == remove:
                items[i] = ""
                count -= 1
                removed = True
                print("Item removed!")
                break
        if not removed:
            print("Item not found!")

    elif choice == "4":
        print("Total items:", count)

    elif choice == "5":
        running = False
        print("Goodbye!")

    else:
        print("Invalid choice, try again.")
```
