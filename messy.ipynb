```python
# messy_shopping.ipynb
print("=== Shopping List Manager ===")

# storing items using separate variables (since no lists allowed)
item1 = ""
item2 = ""
item3 = ""
item4 = ""
item5 = ""

count = 0
running = True

while running:
    print("\n1. Add item")
    print("2. View items")
    print("3. Remove item")
    print("4. Count items")
    print("5. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        if item1 == "":
            item1 = input("Enter item: ")
            count += 1
            print("Item added!")
        elif item2 == "":
            item2 = input("Enter item: ")
            count += 1
            print("Item added!")
        elif item3 == "":
            item3 = input("Enter item: ")
            count += 1
            print("Item added!")
        elif item4 == "":
            item4 = input("Enter item: ")
            count += 1
            print("Item added!")
        elif item5 == "":
            item5 = input("Enter item: ")
            count += 1
            print("Item added!")
        else:
            print("Shopping list is full (max 5 items).")

    elif choice == "2":
        print("Current items:")
        if item1 != "": print(item1)
        if item2 != "": print(item2)
        if item3 != "": print(item3)
        if item4 != "": print(item4)
        if item5 != "": print(item5)

    elif choice == "3":
        remove = input("Enter item to remove: ")
        if remove == item1:
            item1 = ""
            count -= 1
            print("Item removed!")
        elif remove == item2:
            item2 = ""
            count -= 1
            print("Item removed!")
        elif remove == item3:
            item3 = ""
            count -= 1
            print("Item removed!")
        elif remove == item4:
            item4 = ""
            count -= 1
            print("Item removed!")
        elif remove == item5:
            item5 = ""
            count -= 1
            print("Item removed!")
        else:
            print("Item not found!")

    elif choice == "4":
        print("Total items:", count)

    elif choice == "5":
        running = False
        print("Goodbye!")

    else:
        print("Invalid choice, try again.")
```
